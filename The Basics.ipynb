## 1. Elementary functions

### Functions

The domain and image set for the following function is plotted below using Python and matplotlib:

$f(x) = 2x^2 + 6, (x\displaystyle \geq0)$

*The image set is basically the range of output restricted by the domain whilst the domain is a set of permitted input values.*

1. Domain: $(x\displaystyle \geq0)$
2. Image set: $(y\displaystyle \geq 6)$

```python
import numpy as np
import matplotlib.pyplot as plt

# a function that takes arguments x for variable and s for coefficient
def f(x, s):
    return s*x**2 + 6

# choose coefficient s
s = 2   # try positive, negative, or zero

# domain: x >= 0
x = np.linspace(0, 10, 400) # divide 0-10 by 400 for all the x values
y = f(x, s) # a;; the y values

# plot function
plt.plot(x, y, label=f"f(x) = {s}x² + 6")
plt.axhline(6, color="red", linestyle="--", label="y=6 (boundary)") # draw horizontal line on boundary of the function
plt.xlabel("x (domain: x ≥ 0)")
plt.ylabel("f(x)")
plt.title("Function, Domain, and Range")
plt.grid(True)

# Domain shading (x ≥ 0)
plt.axvspan(0, max(x), color="blue", alpha=0.05, label="Domain [0, ∞)")

# Range shading (y-axis)
if s > 0:
    plt.axhspan(6, max(y), color="green", alpha=0.2, label="Range [6, ∞)")
elif s < 0:
    plt.axhspan(min(y), 6, color="green", alpha=0.2, label="Range (-∞, 6]")
else:
    plt.axhline(6, color="green", linewidth=3, alpha=0.5, label="Range {6}")

plt.legend()
plt.show()

# Print domain & range
print("Domain: [0, ∞)")
if s > 0:
    print("Range: [6, ∞)")
elif s < 0:
    print("Range: (-∞, 6]")
else:
    print("Range: {6}")
```

### Inverse Functions

The inverse of the above function (or any function) is simply reversing the mapping i.e. from $x$ to $y$ to $y$ to $x$. Essentially making $x$ the subject in the function: 

$f(x) = 2x^2 + 6, (x\displaystyle \geq0)$.

Re-arranging it becomes: 

$x = \pm \sqrt{\tfrac{1}{2}(y - 6)}$

However, because the domain is $(x\displaystyle \geq0)$ the range of output (the image) will not inlude the negative solutions so the final inverse function is (we change letter of the function from f to g to indicate a separate function):

$g(x) = \sqrt{\tfrac{1}{2}(x - 6)}$

And we can code this in python as follows...




```python
import math

# original function
def g(x):
    return 2 * x**2 + 6

# the inverse of the original function
def g_inverse(x):

    if x < 6:
        print("Undefined for x < 6; needs a complex solution")
    else:
        return math.sqrt((x - 6) /2)

print(f"The original function with an input of 1 returns: {round(g(1),3)}")
print(f"The inverse function with an input of 8 returns back: {round(g_inverse(8),3)}")
    
```

Note that the domain and image also switch around i.e. the image set now becomes the domain and vice versa for the domain 

So:

1. Domain: $(x\displaystyle \geq 6)$
2. Image set: $(y\displaystyle \geq0)$

### Linear functions

One of the more straightforward functions. Shows a direct relationship between two variables typically between $x$ and $y$. They are useful for solving real life problems e.g. cost per unit (variable costs) plus fixed costs can be used to calculate total costs.

The formular is: $f(x) = mx + c$; where $m$ is the gradient (constant rate) and $c$ is the $y$ intercept. In example above $c$ would related to a fixed cost, $m$ to cost per unit and $x$ as the number of units.

You can extrapolate (predict the future especailly overlaying over a scatter plot) or interpolate (look backwards in the past).

Linear functions or linear equations can be combined to solve simultanous linear equations i.e. the solution is the point where the lines of the functions cross for two or more linear functions. They can also be used for optimisation problems such as those used in linear programming.

Consider the following functions (which combined are a system of linear equations):
1. $4x + 4y = -1$
2. $3x + y = 3$

The equations can be solved graphically, algebraically via the Gaussian elimination method or using matrices. The graphically method is illustrated below.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-10, 10, 400)

# Re-arrange the equations in terms of y
y1 = (-1 -4*X) / 4
y2 = 3 - 3*X

plt.plot(X, y1, label="4X + 4y = -1")
plt.plot(X, y2, label="3X + y = 3")

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simultaneous Equations')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

plt.show()
```

The solution for this set of simultanous equations reading off the graph is $x = 1.625$ where $y=-1.875$

The method using matrices is illustrated below:

The system of linear equations are converted to matrix form (*more will be discussed on this in the chapter on Linear Algebra in the other Jupyter Notes*) and solved using the NumPy linear algebra function linalg.

```python
import numpy as np

# Coefficient matrix (left-hand side)
A = np.array([[4, 4],
              [3, 1]])

# Constant matrix (right-hand side)
B = np.array([-1, 3])

# Solve for [x, y]
solution = np.linalg.solve(A, B)

x, y = solution
print(f"x = {x}")
print(f"y = {y}")
```

### The exponential function

The function is really important; sometimes called the natural exponential function because it represents how most things grow in nature e.g. growing bacteria in proposition to it's numbers across time.

It is defined from the Taylor series below.



And all it is is just adding a series of fractions (with factorials on the denominator) plus $1 + x + ....$ to infinity which will only approach the limit of $e^x$.



### The exponential number

The number $e$ itself (which forms part of the exponential function is derived as below)



The second formula can be illustrated with the case of compounding interest. Let's say I have £100 and I deposit it in an bank at 10% per year. At the end of the first year I will have: $£100 \times \left(1 + \frac{10}{100}\right) = £110$. Then on top of that $£110$ (if I don't take it out) I  will have $£112 \times \left(1 + \frac{10}{100}\right) = £121$. To illustrate the point n in the fraction is convinently the same as in the power i.e. the number of periods so for example $\frac{1}{10}$ is 10% for 10 periods (10 years) or $\frac{1}{20}$ is 5% for a period of 20 periods (still 10 years but interest paid twice a year) etc.

There's a repeating pattern here if I do this for 10 years. That pattern is: $(1 + \frac{1}{10}) \times (1 + \frac{1}{10}) \times (1 + \frac{1}{10}) \times.....$ which I can simly re-write as: $(1 + \tfrac{1}{n}\bigr)^{n}$ by factorising out $(1 + \frac{1}{10})$ n times. 

The fraction $\tfrac{1}{n}$ can be broken down into single yearly payments, twice yearly payments or more regular payments over the course of 10 years. The bigger n is the more it appraoches the limit of $e$ or 2.71828.

## Estimation of the exponential function (in Python)

The bigger n (k in the code) becomes the more accurate will get to the limit of the exponential value $e$. The function is called in python using the function from the math library: **math.exp(x)**.

```python
import math

def exp_taylor(x, N=20):
    """
    Approximate e^x by summing the first N+1 terms of its Taylor series.
    
    e^x = sum_{k=0}^∞ x^k / k!
    We truncate at k = N.
    """
    total = 0.0
    term = 1.0      # the k=0 term: x^0 / 0! = 1
    for k in range(N+1):
        total += term
        # prepare next term: term = x^(k+1)/(k+1)!
        term *= x / (k + 1)
    return total

# Example usage:
for x in [1]:
    approx = exp_taylor(x, N=20)
    actual = math.exp(x)
    print(f"x={x:>2}, Taylor≈{approx:.10f}, math.exp={actual:.10f}, error={abs(approx-actual):.2e}")
```

### An important property of the exponential function

Differentiating the exponential function $e^x$ equals itself because:

2. The second line is the original taylor series
3. It is clear that doing the cancellations of the terms in the first line equals the terms in the second line e.g.

$\dfrac{3x^2}{3!}$ and $\dfrac{x^2}{2!}$

### Simplification of logs using log rules

All the simplifications done in python in the following examples make use of the following basic log rules (and also the inverse property of logarithms above):





#### Addition and substraction

```python
import math

# Calculate the original expression: ln(7) + ln(4) - ln(14)
# Note: math.log() computes the natural logarithm (base e)
expression = math.log(7) + math.log(4) - math.log(14)

# Calculate the simplified result: ln(2)
simplified = math.log(2)

print(f"Original Expression: math.log(7) + math.log(4) - math.log(14): {expression}")
print(f"Simplified result (ln 2): {simplified}")
```

#### Multiplication of logs

```python
import sympy

# Tell sympy that x and y are real numbers
x, y = sympy.symbols('x y', real=True)

# Define the expression
expression = sympy.log(sympy.exp(x) * sympy.exp(y))

# Simplify the expression
# We use .expand(log=True) to force it to apply the log rules
simplified_expression = sympy.simplify(expression)

print(f"Original Expression: {expression}")
print(f"Simplified: {simplified_expression}")
```

#### Simplifying logs

```python
from sympy import symbols, exp, log, simplify

# Define the symbolic variable 'x'
x = symbols('x')

# Define the expression: exp(-2 * log(x))
# Note: In SymPy, 'log' represents the natural logarithm (ln)
expression = exp(-2 * log(x))

# Simplify the expression
simplified_expression = simplify(expression)

# Print the original and simplified forms
print(f"Original Expression: e^(-2ln(x))")
print(f"Simplified Expression: {simplified_expression}")
```

The beginning of trigonometry starts with the humble right-angled triangle. All the identities and equations are derived from here:



The trignonometric functions are useful for describing things that repeat over time or in otherwords have some kind of oscillatory nature e.g. a child on a swing, vibrations of atoms in a lattice back and forth around their 'centre' or equilibrium point, bumgee jumping, sound travelling in air etc.

Mathematicians/scientists project these phenomena onto trigonometric functions so as to understand them by modelling.

From the triangle above can be derived the basic trignometric functions of sin, cosine and tangent:

$sin(\theta) = \dfrac ah$, $cos(\theta) = \dfrac bh$, $tan(\theta) = \dfrac ab$

However, these basic functions only have an input range (domain) from 0 to 90 degrees i.e. $0<\theta<\dfrac 2\pi$ because as can be seen on the triangle the angle range can only be extended so far before it doesn't look like a right-angled triangle anymore.

A unit circle can extend the input range (domain):



Also, note that since the trigonometric functions are generally periodic, by going anti-clockwise around the unit circle to point A again we have made a "full 360" in common parlence. This means the basic trigonometric functions can be written as:

$sin(\theta + 2\pi)$ = $sin(\theta)$, $cos(\theta + 2\pi)$ = $cos(\theta)$ i.e. for any $\theta$ this is equal

$tan$ is a bit different as that repeats every $\pi$

Seeing basic trigonometric functions plotted on graphs is easier to digest than the unit circle i.e. it's very easy to see the repeating nature for example $sin(\dfrac 2\pi)$ and $sin(\dfrac{5\pi}{2})$ both have a value of 1

### Inverse Trigonometric functions

Let's consider the case of the inverse of the basic trigonometric functions mentioned above. As already mentioned above inverse functions are found by reversing the mapping i.e. from $x$ to $y$ to $y$ to $x$ i.e. the input changes from $x$ to $y$. 

Taking the sin function as an example:

Say $sin(\dfrac \pi 2)$ = 1 (this can also be seen from the graph of the $sin$ function above) then how can the process be reverse to find $\dfrac \pi 2$? We can switch the axis so that y becomes x and x becomes y i.e. the purple line (with the green line being the original function).

Also, the range of the original $sin$ function is restricted between $-\dfrac \pi 2$ and $\dfrac \pi 2$ because since the $sin$ function repeats for different x values when the x and y axis are swapped around the value of y should only give a unique output (a function is only a function if an input, in the inverse function case y, has only one unique output x); this is why the inverse function needs to be restricted otherwise the input value of 1 will not only return $\dfrac \pi 2$ but also $\dfrac{5\pi}{2}$ and consecutive $\theta +2\pi$ outputs 





The same reasoning is used to derive the inverse functions for $cos(\theta)$ and $tan(\theta)$
