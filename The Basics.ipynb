## Elementary functions

### Functions

The domain and image set for the following function is plotted below using Python and matplotlib:

$f(x) = 2x^2 + 6, (x\displaystyle \geq0)$

*The image set is basically the range of output restricted by the domain whilst the dommain is a set of permitted input values.*

1. Domain: $(x\displaystyle \geq0)$
2. Image set: $(y\displaystyle \geq 6)$

```python
import numpy as np
import matplotlib.pyplot as plt

# a function that takes arguments x for variable and s for coefficient
def f(x, s):
    return s*x**2 + 6

# choose coefficient s
s = 2   # try positive, negative, or zero

# domain: x >= 0
x = np.linspace(0, 10, 400) # divide 0-10 by 400 for all the x values
y = f(x, s) # a;; the y values

# plot function
plt.plot(x, y, label=f"f(x) = {s}x² + 6")
plt.axhline(6, color="red", linestyle="--", label="y=6 (boundary)") # draw horizontal line on boundary of the function
plt.xlabel("x (domain: x ≥ 0)")
plt.ylabel("f(x)")
plt.title("Function, Domain, and Range")
plt.grid(True)

# Domain shading (x ≥ 0)
plt.axvspan(0, max(x), color="blue", alpha=0.05, label="Domain [0, ∞)")

# Range shading (y-axis)
if s > 0:
    plt.axhspan(6, max(y), color="green", alpha=0.2, label="Range [6, ∞)")
elif s < 0:
    plt.axhspan(min(y), 6, color="green", alpha=0.2, label="Range (-∞, 6]")
else:
    plt.axhline(6, color="green", linewidth=3, alpha=0.5, label="Range {6}")

plt.legend()
plt.show()

# Print domain & range
print("Domain: [0, ∞)")
if s > 0:
    print("Range: [6, ∞)")
elif s < 0:
    print("Range: (-∞, 6]")
else:
    print("Range: {6}")
```

The inverse of the above function (or any function) is simply reversing the mapping i.e. from $x$ to $y$ to $y$ to $x$. Essentially making $x$ the subject in the function: 

$f(x) = 2x^2 + 6, (x\displaystyle \geq0)$.

Re-arranging it becomes: 

$x = \pm \sqrt{\tfrac{1}{2}(y - 6)}$

However, because the domain is $(x\displaystyle \geq0)$ the range of output (the image) will not inlude the negative solutions so the final inverse function is (we change letter of the function from f to g to indicate a separate function):

$g(x) = \sqrt{\tfrac{1}{2}(x - 6)}$

And we can code this in python as follows...




```python
import math

# original function
def g(x):
    return 2 * x**2 + 6

# the inverse of the original function
def g_inverse(x):

    if x < 6:
        print("Undefined for x < 6; needs a complex solution")
    else:
        return math.sqrt((x - 6) /2)

print(round(g_inverse(8),3))
    
```

Note that the domain and image also switch around i.e. the image set now becomes the domain and vice versa for the domain 

So:

1. Domain: $(x\displaystyle \geq 6)$
2. Image set: $(y\displaystyle \geq0)$

### Linear functions

One of the more straightforward functions. Shows a direct relationship between two variables typically between $x$ and $y$. They are useful for solving real life problems e.g. cost per unit (variable costs) plus fixed costs can be used to calculate total costs.

The formular is: $f(x) = mx + c$; where $m$ is the gradient (constant rate) and $c$ is the $y$ intercept. In example above $c$ would related to a fixed cost, $m$ to cost per unit and $x$ as the number of units.

You can extrapolate (predict the future especailly overlaying over a scatter plot) or interpolate (look backwards in the past).

Linear functions or linear equations can be combined to solve simultanous linear equations i.e. the solution is the point where the lines of the functions cross for two or more linear functions. They can also be used for optimisation problems such as those used in linear programming.

Consider the following functions (which combined are a system of linear equations):
1. $4x + 4y = -1$
2. $3x + y = 3$

The equations can be solved graphically, algebraically via the Gaussian elimination method or using matrices. The graphically method is illustrated below.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-10, 10, 400)

# Re-arrange the equations in terms of y
y1 = (-1 -4*X) / 4
y2 = 3 - 3*X

plt.plot(X, y1, label="4X + 4y = -1")
plt.plot(X, y2, label="3X + y = 3")

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simultaneous Equations')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

plt.show()
```

The solution for this set of simultanous equations reading off the graph is $x = 2$ where $y=1.666$

The method using matrices is illustrated below:

The system of linear equations are converted to matrix form (*more will be discussed on this in the chapter on Linear Algebra in the other Jupyter Notes*) and solved using the NumPy linear algebra function linalg.

```python
import numpy as np

# Coefficient matrix (left-hand side)
A = np.array([[4, 4],
              [3, 1]])

# Constant matrix (right-hand side)
B = np.array([-1, 3])

# Solve for [x, y]
solution = np.linalg.solve(A, B)

x, y = solution
print(f"x = {x}")
print(f"y = {y}")
```
